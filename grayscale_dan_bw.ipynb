```python
# import library

from PIL import Image
import numpy as np
```

```python
# load citra

img = Image.open("smansa.jpg")
```

```python
# ubah ke grayscali via framework

gray = img.convert('L')
```

```python
gray
```

```python
# definisikan black-white (bw)

bw = np.asarray(gray).copy()
```

```python
# Pixel range is 0...255, 256/2 = 128

bw[bw < 128] = 0    # Black
bw[bw >= 128] = 255 # White
```

```python
# Nhasil bw

imfile = Image.fromarray(bw)
```
